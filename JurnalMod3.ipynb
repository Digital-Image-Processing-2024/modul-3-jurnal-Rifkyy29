```python
import numpy as np
import matplotlib.pyplot as pl
import matplotlib.image as img
import cv2 as cv
```

```python
image = img.imread("IoriYagami.jpg")
image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)

```

```python
x,y= image.shape
imgarray = np.array(image)

lpf_kernel = np.array([0.5, 0.6, 0.7])
hpf_kernel = np.array([-0.5, -0.6, -0.7])

x, y, imgarray
```

```python
def convolution(img, kernel):
    x, y = img.shape
    kernel_height = 1
    kernel_width = 1
    
    padding_height = kernel_height //2
    padding_width = kernel_width //2

    padded_img = np.pad(img, ((padding_height, padding_height), (padding_width, padding_width)), mode='constant', constant_values=0)
    
    result = np.zeros((x, y))
    for i in range(x):
        for j in range(y):
            result[i][j] = np.sum(padded_img[i:i+kernel_height, j:j+kernel_width] * kernel)
    return result
```

```python
decom = convolution(image,lpf_kernel)
aprox = convolution(decom,lpf_kernel)
plt.imshow(aprox,cmap="gray")
```

```python
horizontal = convolution(decom,hpf_kernel)
plt.imshow(horizontal, cmap="gray")
```
